# Regresión con splines

**Veamos algún ejemplo de regresión usando splines cúbicos.**

```r
#install.packages('splines', repos='https://cran.rediris.es/')
```

```r
Boston <- read.csv(file="Boston_data.csv", sep=",", head=TRUE)
```

**La función bs() permite hacer splines polinómicos, donde el número de knots puede especificarse a través de los grados de libertad, df, teniendo en cuenta que df = nº de knots + grado. Donde grado es el grado del polinomio (que es 3 por defecto).**

```r
attach(Boston)
library(splines)
fit = lm(medv~bs(lstat, df=6), data=Boston)
summary(fit)
attr(bs(lstat, df=6), "knots")
```

```r
plot(lstat, medv)
new.data = data.frame(lstat = seq(from = min(lstat), to = max(lstat), length.out = 200))
pred_lm = predict(fit, newdata = new.data, interval="confidence")                      
lines(pred_lm[, 1] ~ new.data$lstat, col = "red")
lines(pred_lm[, 2] ~ new.data$lstat, col = "red", lty="dashed")
lines(pred_lm[, 3] ~ new.data$lstat, col = "red", lty="dashed")
```

**El $R^2$ es algo mejor que el caso que habíamos visto con una dependencia cuadrática (donde obteníamos 0.64). En vez de especificar un número de grados de libertad, que escoge los knots según cuantiles, se puede escoger dónde poner los knots.**

```r
fit2 = lm(medv~bs(lstat, knots=c(8, 15, 20)), data=Boston)
summary(fit2)
```

```r
plot(lstat, medv)
pred_lm2 = predict(fit2, newdata = new.data)                      
lines(pred_lm2~ new.data$lstat, col = "red")
```

**También se pueden usar splines naturales, que imponen una dependencia lineal en los extremos (p.ej. polinomio de 2º orden en vez de 3º), para evitar un comportamiento errático en esas zonas. En este caso el df es el nº de knots más 1.**

```r
fit3 = lm(medv~ns(lstat, df=4), data=Boston)
pred_lm3 = predict(fit3, newdata=new.data)         
plot(lstat, medv)
#lines(pred_lm3~ new.data$lstat, col = "red")
lines(new.data$lstat[1:29], pred_lm3[1:29],col=2, lwd=3)
lines(new.data$lstat[30:53], pred_lm3[30:53],col=3, lwd=3)
lines(new.data$lstat[54:84], pred_lm3[54:84],col=4, lwd=3)
lines(new.data$lstat[85:200], pred_lm3[85:200],col=5, lwd=3)
```
