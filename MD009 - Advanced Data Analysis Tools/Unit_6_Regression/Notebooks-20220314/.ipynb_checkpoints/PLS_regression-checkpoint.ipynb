# Ejemplo de regresión con Partial Least Squares (PLS)

**Utilizaremos el mismo dataset de los otros ejemplos.**

```r
Boston <- read.csv(file="Boston_data.csv", sep=",", head=TRUE)
attach(Boston)
```

**Realizamos una regresión PLS con 10 componentes y Leave-One-Out cross-validation:** 

```r
library(pls)
plsr.fit = plsr(medv~., ncomp=10, data=Boston, validation="LOO")
summary(plsr.fit)
```

```r
#Root Mean Square Error of Predition
plot(RMSEP(plsr.fit), legendpos = "topright")
```

**Score plots de los 3 primeros componentes:**

```r
plot(plsr.fit, plottype = "scores", comps = 1:3)
```

**Comparemos el resultado de escoger 4 o 10 componentes:**

```r
plsr.fit2 = plsr(medv~., ncomp=4, data=Boston)
```

```r
plot(plsr.fit2)
abline(0, 1, col = "red")
```

```r
plot(plsr.fit)
abline(0, 1, col = "red")
```
