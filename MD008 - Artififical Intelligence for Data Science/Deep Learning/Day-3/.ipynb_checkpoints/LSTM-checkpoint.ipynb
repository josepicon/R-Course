```python
# Packages 
import numpy as np
import tensorflow as tf

from tensorflow.keras import layers, models
from pandas import concat
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
```

```python
# Create our vector and flatten it to reduce dimensionality
vector = np.array([i*np.random.normal(1, 0.2) + np.random.random(1) for i in range(50)])
vector = vector.flatten()

# reshape for fitting LSTM
vec = np.reshape(vector, (1, 50, 1))
lstm = tf.keras.layers.LSTM(50)
output = lstm(vec)

# Visualize the subset
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 5))
fig.suptitle('Applying Long Short Term Memory')
ax1.plot(vector, label = 'Original vector')
ax1.legend()
ax1.set_title('Original vector')

ax2.plot(vector, label = 'Original vector')
ax2.plot(output[0,:], label = 'LSTMed vector')
ax2.legend()
ax2.set_title('Comparison')
plt.show()
```

## Predicting model

```python
model = models.Sequential()
model.add(tf.keras.layers.Conv1D(filters=1,
                                 kernel_size=3,
                                 input_shape=(50, 1)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(1))

model.compile(optimizer='sgd', loss='mse')

following_value = model(vec)
print('Forecasted value:', following_value[0,0].numpy())
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return np.array(X), np.array(y)
 
# define input sequence
raw_seq = vector

# choose a number of time steps (the bigger, the longer-term knowledge will maintain but it'll also increase computational)
n_steps = 5

# split into samples
X, y = split_sequence(raw_seq, n_steps)

# Reshape for fitting the network
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))

# define model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.LSTM(50, activation='relu', input_shape=(n_steps, n_features)))
model.add(tf.keras.layers.Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=50, verbose=0)
```

```python
pred = []
x_pred = []
# Predicting some of the last samples
for i in range(40, 51):
  low_lim = i-5
  up_lim = i
  x_pred.append(i-1)
  x_input = vector[low_lim:up_lim]
  x_input = x_input.reshape((1, n_steps, n_features))
  yhat = model.predict(x_input, verbose=0)
  pred.append(yhat[0][0])
```

## LSTM forecasting

```python
plt.figure(figsize = (15, 5))
plt.plot(vector, label = 'original vector')
plt.plot(x_pred, pred, label = 'predictions')
plt.show()
```

### CNN forecasting

```python
model_cnn = tf.keras.models.Sequential()
model_cnn.add(tf.keras.layers.Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_steps, n_features)))
model_cnn.add(tf.keras.layers.MaxPooling1D(pool_size=2))
model_cnn.add(tf.keras.layers.Dense(50, activation='relu'))
model_cnn.add(tf.keras.layers.Dense(1))
model_cnn.compile(optimizer='adam', loss='mse')
```

```python
model_cnn.fit(X, y, epochs=50, verbose=0)
```

```python
pred = []
x_pred = []
# Predicting some of the last samples
for i in range(40, 51):
  low_lim = i-5
  up_lim = i
  x_pred.append(i-1)
  x_input = vector[low_lim:up_lim]
  x_input = x_input.reshape((1, n_steps, n_features))
  yhat = model_cnn.predict(x_input, verbose=0)
  pred.append(yhat[0][0])
```

```python
plt.figure(figsize = (15, 5))
plt.plot(vector, label = 'original vector')
plt.plot(x_pred, pred, label = 'predictions')
plt.show()
```

```python
# Outlier cell
pred2 = [value[0] for value in pred]

maes = np.abs(vector[-11:] - pred2)

max_mae = np.max(maes)

print('Threshold error for Outlier detection:', max_mae)
```
